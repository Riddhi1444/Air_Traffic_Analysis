### How can we handle missing values and anomalies in the dataset?

```python
df.isnull().sum()
```

The data was loaded and was checked for any null values.
We found that there are null values present in "Operating Airline IATA Code" and "Published Airline IATA Code" column.

In this the null values as filled with "unknown" value.

```python
df.isnull().sum()
```

## Exploratory Data Analysis (EDA):

### What are the key statistics for passenger counts, and how are they distributed?

```python
df['Passenger Count'].describe()
```

The difference between the mean and median(50%) indicates the presence of outliers in the data such as max value(65k) which makes the data right-skewed.

### Which regions (GEO Summary and GEO Region) have the highest and lowest passenger traffic?

```python
df.groupby('GEO Region')['Passenger Count'].sum().nlargest(5)
```

```python
df.groupby('GEO Region')['Passenger Count'].sum().nsmallest(5)
```

"US" has the highest passenger traffic followed by "Asia" while the least traffic was found in "South America" followed by "Central America". 

### How do the different activity types (Deplaned, Enplaned, etc.) contribute to passenger counts?

```python
sb.countplot(x=df['Activity Type Code'])
plt.show()
```

It was found that activities like "Deplaned" and "Enplaned" contributed more to the passenger count that "Thru/Transit".

## Temporal Analysis:

### How has passenger traffic evolved over the years and months?

```python
pivot=pd.pivot_table(data=df,index='Year',columns='Month',values='Passenger Count',aggfunc='sum')
pivot
```

It can be clearly concluded that most of the traffic is in "August" and "July". And there is a sudden leap in the passsenger count after 2012. 

### Are there specific months or seasons that show peak or low passenger traffic?

```python
plt.figure(figsize=(20,9))
sb.lineplot(x=df['Month'],y=df['Passenger Count'])
plt.xlabel("Months")
plt.ylabel("Passenger Count")
plt.show()
```

"August,June,July" have most of the passengers while people travelled least in "February".

### What trends can be identified from year-over-year passenger data?

```python
plt.figure(figsize=(20,9))
sb.heatmap(pivot,annot=True)
plt.title("Trends for year-over-year passenger data")
plt.xlabel("Months")
plt.ylabel("Years")
```

A sudden increase in passenger count was noticable froom 2012. 

## Airline Analysis:

### Which airlines handle the highest and lowest number of passengers?

```python
df.groupby('Operating Airline')['Passenger Count'].sum().nlargest(5)
```

```python
df.groupby("Operating Airline")["Passenger Count"].sum().nsmallest(5)
```

"United Airlines" has the highest number of passenger while "Evergreen International Airlines" has the lowest number of passengers.There is a extreme diffence between the two. 

### How is passenger traffic distributed among different airlines?

```python
plt.figure(figsize=(20,9))
sb.barplot(x=df['Operating Airline'],y=df['Passenger Count'])
plt.xticks(rotation=90)
plt.show()
```

"American Airlines" has most of the passengers, other airline such as "Delta Airlines","US Airways","United Airlines","Southwest Airlines","virgin Airlines" capture rest of the market while "Evergreen International Airlines" and some others has almost no passengers. 

## Terminal and Boarding Area Analysis:

### Which terminals and boarding areas are the busiest in terms of passenger traffic?

```python
df.groupby('Terminal')['Passenger Count'].sum().nlargest()
```

"Terminal 3" is the busiest terminal followed by "International".

```python
df.groupby('Boarding Area')['Passenger Count'].sum().nlargest()
```

"F" is the busiest boarding area followed by "B". 

### Is there a notable difference in usage between terminals or boarding areas?

```python
x=df.groupby(['Terminal','Boarding Area'])['Passenger Count'].sum().reset_index().sort_values('Passenger Count',ascending=False)
x
```

```python
plt.figure(figsize=(20,9))
sb.barplot(x=x['Terminal'],y=x['Passenger Count'],hue=x['Boarding Area'])
plt.show()
```

Some Terminals and Borading Areas like "Terminal 3","F" and "B" are used much more often compared to others.
